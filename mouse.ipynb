```python
import pandas as pd
pd.set_option('display.max_columns', None)
data=pd.read_csv('./vdjdb.txt',sep='\t', header=0)

data
```

```python
columns_to_drop = ['reference.id', 'method', 'meta','cdr3fix','web.method','web.method.seq','web.cdr3fix.nc','web.cdr3fix.unmp','mhc.a','mhc.b','mhc.class']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
# Filter by species
data_human = data[data['species'] == 'HomoSapiens']
data_mouse = data[data['species'] == 'MusMusculus']
data_monkey = data[data['species'] == 'MacacaMulatta'] 
#next analysis is for mouse only
df=data_mouse
df
```

```python
from tcrdist.repertoire import TCRrep
import pandas as pd
# 针对alpha链和beta链的数据进行筛选
data_alpha = df[df['gene'] == 'TRA'].copy()
data_beta = df[df['gene'] == 'TRB'].copy()
# 将数据列重命名以匹配TCRrep的期望格式
data_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene'}, inplace=True)
data_beta.rename(columns={'cdr3': 'cdr3_b_aa', 'v.segm': 'v_b_gene', 'j.segm': 'j_b_gene'}, inplace=True)
data_alpha.to_csv('data_alpha.csv')
data_beta.to_csv('beta.csv')
data_beta
```

```python
import numpy as np

#创造‘TCRrep’类
#alpha chain
tr_alpha = TCRrep(cell_df=data_alpha, 
                  organism='mouse', 
                  chains=['alpha'], 
                  db_file='alphabeta_gammadelta_db.tsv',
                  deduplicate=True)
tr_alpha.compute_distances()

alpha_distance_matrix = tr_alpha.pw_alpha
np.save('alpha_distance_mouse_matrix.npy', alpha_distance_matrix)
tr_alpha.pw_alpha
```

```python
#beta chain
from tcrdist.repertoire import TCRrep
import numpy as np
data_beta = pd.read_csv("beta.csv")
tr_beta = TCRrep(cell_df=data_beta, 
                 organism='mouse', 
                 chains=['beta'], 
                 db_file='alphabeta_gammadelta_db.tsv',
                 deduplicate=True)
tr_beta.compute_distances()
beta_distance_matrix = tr_beta.pw_beta
np.save('beta_distance_mouse_matrix.npy', beta_distance_matrix)
tr_beta.pw_beta
```

```python
# 通过complex.id进行合并，假设每个complex.id对应一对alpha和beta链
# 确保complex.id不等于0
data_alpha = data_alpha[data_alpha['complex.id'] != 0]
data_beta = data_beta[data_beta['complex.id'] != 0]
# 合并alpha和beta链数据（拼接）
data_combined = pd.merge(data_alpha, data_beta, on='complex.id', suffixes=('_alpha', '_beta'))
data_combined['combined_feature'] = data_combined['cdr3_a_aa'] + "_" + data_combined['cdr3_b_aa']

data_combined.to_csv('data_combined.csv')
data_combined
```

```python
data_combined['cdr3_b_aa'] = data_combined['combined_feature']
tr = TCRrep(cell_df = data_combined,
            organism = 'mouse', 
            chains = ['beta'],  # 这里chains参数是必须的，但因为我们用的是综合特征，所以实际选择哪个链并不重要
            db_file='alphabeta_gammadelta_db.tsv', 
            )
tr.compute_distances()
tr.pw_beta
```

```python
distance_matrix = tr.pw_beta
np.save('combined_distance_mouse_matrix.npy', distance_matrix)
```
