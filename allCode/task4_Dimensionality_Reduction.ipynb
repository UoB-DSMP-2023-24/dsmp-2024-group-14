```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Load distance matrix
distance_matrix = np.load('alpha_distance_human_matrix.npy')

# Load the dataset containing sequence features
data = pd.read_csv("human_alpha.csv")

# Extract features (antigen.epitope)
categories = data['epitope']

# Perform dimensionality reduction using t-SNE
tsne = TSNE(n_components=2, metric='precomputed', init='random',random_state=42 )
embeddings = tsne.fit_transform(distance_matrix)

# Get unique categories
unique_categories = np.unique(categories)

# Plot the embeddings with different colors for different categories
plt.figure(figsize=(8, 6))
for category in unique_categories:
    indices = (categories == category)
    plt.scatter(embeddings[indices, 0], embeddings[indices, 1], label=category, s=20)

#plt.title('Visualization of human alpha TCRs in 2 Dimensions Colored by antigen.epitope')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Place legend on the right
plt.show()
```
