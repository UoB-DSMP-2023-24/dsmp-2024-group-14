```python
import pandas as pd
pd.set_option('display.max_columns', None)
data=pd.read_csv('./vdjdb.txt',sep='\t', header=0)
data
```

```python
columns_to_drop = ['reference.id', 'method', 'meta','cdr3fix','web.method','web.method.seq','web.cdr3fix.nc','web.cdr3fix.unmp','mhc.a','mhc.b','mhc.class','antigen.gene']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
data = data[data['vdjdb.score'] != 0]
data
```

```python
data.isnull().sum()
```

```python
data=data.dropna() # dropping rows with any column having null values
```

```python
data
```

```python
data['antigen.epitope'].value_counts()
```

```python
duplicate_rows = data[data.duplicated()]
print(duplicate_rows)
```

```python
data = data.drop_duplicates()

print(data)
```

```python
data
```

```python
data['species'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.hist(data['species'], bins=10)  
plt.xlabel('species')
plt.ylabel('Frequency')
plt.title('Histogram of species')
plt.show()
```

```python
# Filter by species
data_human = data[data['species'] == 'HomoSapiens']
data_mouse = data[data['species'] == 'MusMusculus']
#data_monkey = data[data['species'] == 'MacacaMulatta']
df1=data_human
df2=data_mouse
```

# human

```python
# Filter data for alpha chain and beta chain
df1_alpha = df1[df1['gene'] == 'TRA'].copy()
df1_beta = df1[df1['gene'] == 'TRB'].copy()
```

#  human alpha

```python
df1_alpha
```

```python
duplicate_rows = df1_alpha[df1_alpha.duplicated(subset=['cdr3', 'v.segm','j.segm', 'antigen.epitope'])]
print(duplicate_rows)
```

```python
#Drop duplicate TCRs
df1_alpha.drop(duplicate_rows.index, inplace=True)
print(df1_alpha)
```

```python
n_orig = df1_alpha.shape[0]
print(f'The dataset contains {df1_alpha.shape[0]} TCRs for {len(df1_alpha["antigen.epitope"].unique())} different epitopes. The amount of TCRs per epitope:\n')
df1_alpha["antigen.epitope"].value_counts()
```

```python
#Drop epitopes with less than 20 TCRs
peptide_counts = df1_alpha['antigen.epitope'].value_counts()
peptide_list = peptide_counts[peptide_counts >=20].index
df1_alpha = df1_alpha[df1_alpha['antigen.epitope'].isin(peptide_list)]
df1_alpha.reset_index(drop=True, inplace=True)

print(f'The new amount of TCRs per epitope, after dropping epitopes with less than 20 TCRs:\n')
df1_alpha["antigen.epitope"].value_counts()
```

```python
df1_alpha
```

```python
df1_alpha['vdjdb.score'].value_counts()
```

```python
# Rename data columns to match TCRrep's expected format
df1_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene', 'antigen.epitope': 'epitope'}, inplace=True)

df1_alpha.to_csv('human_alpha.csv')
```

# human combined alpha and beta chain

```python
df1_alpha1 = df1[df1['gene'] == 'TRA'].copy()
df1_beta1 = df1[df1['gene'] == 'TRB'].copy()
```

```python
# Merge by complex.id, assuming that each complex.id corresponds to a pair of alpha and beta chains
# Make sure complex.id is not equal to 0
df1_alpha1 = df1_alpha1[df1_alpha1['complex.id'] != 0]
df1_beta1 = df1_beta1[df1_beta1['complex.id'] != 0]
# Merge alpha and beta chain data
common_columns = ['species', 'antigen.epitope','antigen.species','vdjdb.score']
df1_beta1_reduced = df1_beta1.drop(common_columns, axis=1)
human_combined = pd.merge(df1_alpha1,df1_beta1_reduced, on='complex.id' )
```

```python
human_combined 
```

```python
duplicate_rows = human_combined[human_combined.duplicated(subset=['cdr3_x', 'v.segm_x','j.segm_x','cdr3_y', 'v.segm_y','j.segm_y', 'antigen.epitope'])]
print(duplicate_rows)
```

```python
#Drop duplicate TCRs
human_combined.drop(duplicate_rows.index, inplace=True)
print(human_combined)
```

```python
n_orig = human_combined.shape[0]
print(f'The dataset contains {human_combined.shape[0]} TCRs for {len(human_combined["antigen.epitope"].unique())} different epitopes. The amount of TCRs per epitope:\n')
human_combined["antigen.epitope"].value_counts()
```

```python
#Drop epitopes with less than 20 TCRs
peptide_counts = human_combined['antigen.epitope'].value_counts()
peptide_list = peptide_counts[peptide_counts >=20].index
human_combined = human_combined[human_combined['antigen.epitope'].isin(peptide_list)]
human_combined.reset_index(drop=True, inplace=True)

print(f'The new amount of TCRs per epitope, after dropping epitopes with less than 20 TCRs:\n')
human_combined["antigen.epitope"].value_counts()
```

```python
human_combined
```

```python
human_combined['vdjdb.score'].value_counts()
```

```python
human_combined.rename(columns={'cdr3_x': 'cdr3_a_aa', 'v.segm_x': 'v_a_gene', 'j.segm_x': 'j_a_gene', 'cdr3_y': 'cdr3_b_aa', 'v.segm_y': 'v_b_gene', 'j.segm_y': 'j_b_gene', 'antigen.epitope': 'epitope'}, inplace=True)

human_combined.to_csv('human_combined.csv')
```
