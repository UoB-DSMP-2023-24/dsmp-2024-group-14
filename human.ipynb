```python
import pandas as pd
pd.set_option('display.max_columns', None)
data=pd.read_csv('./vdjdb.txt',sep='\t', header=0)

data
```

```python
columns_to_drop = ['reference.id', 'method', 'meta','cdr3fix','web.method','web.method.seq','web.cdr3fix.nc','web.cdr3fix.unmp','mhc.a','mhc.b','mhc.class']
data.drop(columns=columns_to_drop, inplace=True)
data=data.dropna()
data
```

```python
data=data[data['vdjdb.score'] != 0]
data
```

```python
# Filter by species
data_human = data[data['species'] == 'HomoSapiens']
data_mouse = data[data['species'] == 'MusMusculus']
data_monkey = data[data['species'] == 'MacacaMulatta'] 
```

```python
#next analysis is for humans 
df=data_human
df
```

```python
data_human['antigen.epitope'].value_counts()
```

```python
from tcrdist.repertoire import TCRrep
import pandas as pd
# # Filter data for alpha chain and beta chain
data_alpha = df[df['gene'] == 'TRA'].copy()
data_beta = df[df['gene'] == 'TRB'].copy()
# Rename data columns to match TCRrep's expected format
data_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene', 'antigen.epitope': 'epitope'}, inplace=True)
data_beta.rename(columns={'cdr3': 'cdr3_b_aa', 'v.segm': 'v_b_gene', 'j.segm': 'j_b_gene', 'antigen.epitope': 'epitope'}, inplace=True)
data_alpha.to_csv('human_alpha.csv')
data_beta.to_csv('human_beta.csv')
```

```python
import numpy as np
#alpha chain
tr_alpha = TCRrep(cell_df=data_alpha, 
                  organism='human', 
                  chains=['alpha'], 
                  db_file='alphabeta_gammadelta_db.tsv',
                  deduplicate=True)
#tr_alpha.compute_distances()

alpha_distance_matrix = tr_alpha.pw_alpha
alpha_distance_matrix
```

```python
def min_max_normalize(dist_matrix):
    min_val = np.min(dist_matrix)
    max_val = np.max(dist_matrix)
    norm_matrix = (dist_matrix - min_val) / (max_val - min_val)
    return norm_matrix

normalized_alpha_distances_matrix = min_max_normalize(alpha_distance_matrix)

print(normalized_alpha_distances_matrix)
np.save('alpha_distance_human_matrix.npy', normalized_alpha_distances_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
matrix = normalized_alpha_distances_matrix

sns.heatmap(matrix, cmap='viridis')
plt.title("TCR Human Alpha Chain Distance Matrix")
plt.show()
```

```python
from tcrdist.repertoire import TCRrep
import numpy as np
tr_beta = TCRrep(cell_df=data_beta, 
                 organism='human', 
                 chains=['beta'], 
                 db_file='alphabeta_gammadelta_db.tsv',
                 deduplicate=True)
beta_distance_matrix = tr_beta.pw_beta
beta_distance_matrix
```

```python
normalized_beta_distances_matrix= min_max_normalize(beta_distance_matrix)
print(normalized_beta_distances_matrix)
np.save('beta_distance_human_matrix.npy', normalized_beta_distances_matrix)
```

```python
import seaborn as sns

matrix = normalized_beta_distances_matrix

sns.heatmap(matrix, cmap='viridis')
#plt.imshow(matrix, cmap='viridis')
plt.title("TCR Human Beta Chain Distance Matrix")
#plt.colorbar()
plt.show()
```

```python
# Merge by complex.id, assuming that each complex.id corresponds to a pair of alpha and beta chains
# Make sure complex.id is not equal to 0
data_alpha = data_alpha[data_alpha['complex.id'] != 0]
data_beta = data_beta[data_beta['complex.id'] != 0]
# Merge alpha and beta chain data
common_columns = ['species', 'epitope','antigen.species','antigen.gene']
data_beta_reduced = data_beta.drop(common_columns, axis=1)
data_combined = pd.merge(data_alpha,data_beta_reduced, on='complex.id' )
data_beta.to_csv('human_combined.csv')
data_combined
```

```python
#data_combined= pd.read_csv("human_combined.csv")
tr = TCRrep(cell_df=data_combined, 
                  organism='human', 
                   chains = ['alpha', 'beta'],
                  db_file='alphabeta_gammadelta_db.tsv')
# Calculate the distance between alpha and beta chains
tr.compute_distances()

# Separate distance matrices for alpha and beta chains
alpha_distances_matrix = tr.pw_alpha
beta_distances_matrix = tr.pw_beta

# Combine the distance _matrix of alpha and beta chains
combined_distances_matrix = alpha_distances_matrix + beta_distances_matrix
combined_distances_matrix
```

```python
normalized_alpha_distances_matrix = min_max_normalize(alpha_distances_matrix)
normalized_beta_distances_matrix= min_max_normalize(beta_distances_matrix)
normalized_combined_distances_matrix=min_max_normalize(combined_distances_matrix)
print(normalized_combined_distances_matrix )
```

```python
import seaborn as sns

matrix = normalized_combined_distances_matrix

sns.heatmap(matrix, cmap='viridis')
plt.title("TCR Human Combined Chain Distance Matrix")
plt.show()
```
