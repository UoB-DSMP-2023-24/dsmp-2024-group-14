```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
data=pd.read_csv('./vdjdb.txt',sep='\t', header=0)

data
```

```python
columns_to_drop = ['reference.id', 'method', 'meta','cdr3fix','web.method','web.method.seq','web.cdr3fix.nc','web.cdr3fix.unmp','mhc.a','mhc.b','mhc.class']
data.drop(columns=columns_to_drop, inplace=True)
data=data.dropna()
data=data[data['vdjdb.score'] != 0]
data
```

```python
# Filter by species
data_human = data[data['species'] == 'HomoSapiens']
data_mouse = data[data['species'] == 'MusMusculus']
data_monkey = data[data['species'] == 'MacacaMulatta'] 
#next analysis is for humans 
df=data_human
```

```python
from tcrdist.repertoire import TCRrep
import pandas as pd
# 针对alpha链和beta链的数据进行筛选
data_alpha = df[df['gene'] == 'TRA'].copy()
data_beta = df[df['gene'] == 'TRB'].copy()
# 将数据列重命名以匹配TCRrep的期望格式
data_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene'}, inplace=True)
data_beta.rename(columns={'cdr3': 'cdr3_b_aa', 'v.segm': 'v_b_gene', 'j.segm': 'j_b_gene'}, inplace=True)
data_alpha
```

alpha

```python
#alpha chain
tr_alpha = TCRrep(cell_df=data_alpha, 
                  organism='human', 
                  chains=['alpha'], 
                  db_file='alphabeta_gammadelta_db.tsv',
                  deduplicate=True)
tr_alpha.compute_distances()

alpha_distance_matrix = tr_alpha.pw_alpha
np.save('alpha_distance_human_matrix.npy', alpha_distance_matrix)


def min_max_normalize(dist_matrix):
    min_val = np.min(dist_matrix)
    max_val = np.max(dist_matrix)
    norm_matrix = (dist_matrix - min_val) / (max_val - min_val)
    return norm_matrix
normalized_alpha_distances = min_max_normalize(alpha_distance_matrix)
normalized_alpha_distances
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.imshow(normalized_alpha_distances, cmap='viridis', interpolation='nearest')
plt.title("TCR human alpha Chain Pairwise Distance Matrix")
plt.colorbar()
plt.show()

```

Beta

```python
#beta chain
tr_alpha = TCRrep(cell_df=data_beta, 
                  organism='human', 
                  chains=['beta'], 
                  db_file='alphabeta_gammadelta_db.tsv',
                  deduplicate=True)
tr_alpha.compute_distances()

beta_distance_matrix = tr_alpha.pw_beta
np.save('alpha_distance_human_matrix.npy', beta_distance_matrix)
normalized_beta_distances = min_max_normalize(beta_distance_matrix)
normalized_beta_distances
```

```python
plt.imshow(normalized_beta_distances, cmap='viridis', interpolation='nearest')
plt.title("TCR beta Chain Pairwise Distance Matrix")
plt.colorbar()
plt.show()
```

combined:

```python
# 通过complex.id进行合并，假设每个complex.id对应一对alpha和beta链
# 确保complex.id不等于0
data_alpha = data_alpha[data_alpha['complex.id'] != 0]
data_beta = data_beta[data_beta['complex.id'] != 0]
# 合并alpha和beta链数据（拼接）
common_columns = ['species', 'antigen.epitope','antigen.species','antigen.gene']
data_beta_reduced = data_beta.drop(common_columns, axis=1)
data_combined = pd.merge(data_alpha,data_beta_reduced, on='complex.id' )
data_combined
```

```python
tr = TCRrep(cell_df=data_combined, 
                  organism='human', 
                   chains = ['alpha', 'beta'],
                  db_file='alphabeta_gammadelta_db.tsv')
# 计算 alpha 和 beta 链的距离
tr.compute_distances()

# 你现在可以访问 alpha 和 beta 链的单独距离矩阵
alpha_distances = tr.pw_alpha
beta_distances = tr.pw_beta

# 结合 alpha 和 beta 链的距离
combined_distances = alpha_distances + beta_distances


# 假设 dist_matrix 是你的距离矩阵
normalized_alpha_distances = min_max_normalize(alpha_distances)
normalized_beta_distances= min_max_normalize(beta_distances)
normalized_combined_distances=min_max_normalize(combined_distances)


print(normalized_combined_distances )
```

```python

plt.imshow(normalized_combined_distances , cmap='viridis', interpolation='nearest')
plt.title("TCR combined Chain Pairwise Distance Matrix")
plt.colorbar()
plt.show()
```
