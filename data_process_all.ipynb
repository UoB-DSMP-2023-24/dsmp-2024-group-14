```python
import pandas as pd
pd.set_option('display.max_columns', None)
data=pd.read_csv('./vdjdb.txt',sep='\t', header=0)
data
```

```python
columns_to_drop = ['reference.id', 'method', 'meta','cdr3fix','web.method','web.method.seq','web.cdr3fix.nc','web.cdr3fix.unmp','mhc.a','mhc.b','mhc.class','antigen.gene']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
# Filter by species
data_human = data[data['species'] == 'HomoSapiens']
data_mouse = data[data['species'] == 'MusMusculus']
#data_monkey = data[data['species'] == 'MacacaMulatta']
df1=data_human
df2=data_mouse
```

```python
# # Filter data for alpha chain and beta chain
df1_alpha = df1[df1['gene'] == 'TRA'].copy()
df1_beta = df1[df1['gene'] == 'TRB'].copy()
# Rename data columns to match TCRrep's expected format
#data_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene', 'antigen.epitope': 'epitope'}, inplace=True)
#data_beta.rename(columns={'cdr3': 'cdr3_b_aa', 'v.segm': 'v_b_gene', 'j.segm': 'j_b_gene', 'antigen.epitope': 'epitope'}, inplace=True)
```

```python
df1_alpha
```

```python
duplicate_rows = df1_alpha[df1_alpha.duplicated(subset=['cdr3', 'v.segm','j.segm', 'antigen.epitope'])]
print(duplicate_rows)
```

```python
#Drop duplicate TCRs
df1_alpha.drop(duplicate_rows.index, inplace=True)
print(df1_alpha)
```

```python
n_orig = df1_alpha.shape[0]
print(f'The dataset contains {df1_alpha.shape[0]} TCRs for {len(df1_alpha["antigen.epitope"].unique())} different epitopes. The amount of TCRs per epitope:\n')
df1_alpha["antigen.epitope"].value_counts()
```

```python
#Drop epitopes with less than 100 TCRs
peptide_counts = df1_alpha['antigen.epitope'].value_counts()
peptide_list = peptide_counts[peptide_counts >=100].index
df1_alpha = df1_alpha[df1_alpha['antigen.epitope'].isin(peptide_list)]
df1_alpha.reset_index(drop=True, inplace=True)

print(f'The new amount of TCRs per epitope, after dropping epitopes with less than 100 TCRs:\n')
df1_alpha["antigen.epitope"].value_counts()
```

```python
df1_alpha
```

```python
#new_data=pd.read_csv('human_combined.csv')
score_distribution = pd.pivot_table(df1_alpha, index='antigen.epitope', columns='vdjdb.score', aggfunc='size', fill_value=0)
print(score_distribution)
```

```python
category_counts = df1_alpha['vdjdb.score'].value_counts(normalize=True)

print("每个类别所占比例:")
print(category_counts)
```

```python
import pandas as pd
# 计算score为0的数据总数
score_0_count = df1_alpha[df1_alpha['vdjdb.score'] == 0].shape[0]
# 保留score为1、2、3的数据
other_scores = df1_alpha[df1_alpha['vdjdb.score'] != 0]
score_other_count = other_scores.shape[0]

# 计算目标数量(50%的score为0的数据总数)
target_count = int(score_other_count)
print(target_count)

# 对score为0的数据进行分层抽样
score_0_data = df1_alpha[df1_alpha['vdjdb.score'] == 0]
epitope_groups = score_0_data.groupby('antigen.epitope')

# 计算每个epitope组的样本数量
epitope_sizes = epitope_groups.size()
min_size = epitope_sizes.min()
sample_size = min(min_size, int(target_count / len(epitope_sizes)))
print(sample_size)

sampled_groups = []
for epitope, group in epitope_groups:
    sampled_group = group.sample(n=sample_size, random_state=42)
    sampled_groups.append(sampled_group)

sampled_score_0 = pd.concat(sampled_groups, ignore_index=True)


# 合并抽样结果
sampled_data = pd.concat([sampled_score_0, other_scores])

print(f"Original dataset size: {len(df1_alpha)}")
print(f"Sampled dataset size: {len(sampled_data)}")
print(sampled_data['vdjdb.score'].value_counts(normalize=True))
```

```python
score_distribution = pd.pivot_table(sampled_data, index='antigen.epitope', columns='vdjdb.score', aggfunc='size', fill_value=0)
print(score_distribution)
```

```python
df1_alpha = sampled_data["antigen.epitope"].value_counts()
df1_alpha = sampled_data
```

```python
df1_alpha.rename(columns={'cdr3': 'cdr3_a_aa', 'v.segm': 'v_a_gene', 'j.segm': 'j_a_gene', 'antigen.epitope': 'epitope'}, inplace=True)

df1_alpha.to_csv('human_alpha.csv')
```
